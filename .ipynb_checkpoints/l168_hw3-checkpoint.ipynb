## HW Question 2.4

**We can do better than impressionistic evaluation!** Remember, MFCCs are just matrices of numbers, which is just why neural networks love them. They give us a nice way to evaluate the similarity of matrices derived from sound. Generally, speech processing uses fancier approaches like HMMs or CNNs, but we're going to do this in a very boring, very easy way: using the **Frobenius norm** of the two matrices.

We can think about this as being akin to taking the euclidean distance between each pair of points in the two matrices, and taking the sum of squared distances. This gives us "How far apart are these matrices in space", effectively, and smaller numbers mean more similar.

Below is a chunk of code which creates the MFCCs for three files, and computes the frobenius norms between `pure` `pulse` and `warble`:

```python
# Adjust the length of the pulse tone to match the length of the other two, as we need matrices of equal size.
s800pulse_datatrimmed = s800pulse_data[:22050]
# Create MFCCs
spulse_mfcc = librosa.feature.mfcc(y=s800pulse_datatrimmed, sr=s800pulse_sr,n_mfcc=11)
swarble_mfcc = librosa.feature.mfcc(y=s800warble_data, sr=s800warble_sr,n_mfcc=11)
spure_mfcc = librosa.feature.mfcc(y=s800pure_data, sr=s800pure_sr,n_mfcc=11)
# Compute the norms between pure and pulse
purepulsenorm = np.linalg.norm(spure_mfcc - spulse_mfcc, 'fro')
print(purepulsenorm)
# Compute the norms between pure and warble
purewarblenorm = np.linalg.norm(spure_mfcc - swarble_mfcc, 'fro')
print(purewarblenorm)
```

We can see that the frobenius norm (representing the difference in sounds) is *much* smaller between the warble and pure tones, indicating that those sounds are much closer. If you'd like, you can plot the MFCCs yourself to see why.
